```python
import pandas as pd
import glob
```

```python
csvs = glob.glob('tw_future/*_fut.csv')
csvs
```

```python
df = pd.concat([pd.read_csv(csv, encoding='big5', low_memory=False, dtype=object) for csv in csvs])
df.columns
```

```python
cols = ['Date', 'Contract', 'Expiration', 'Open', 'High', 'Low', 'Close', 'Volume',
        'Change', 'PctChange', 'Settlement', 'OpenInterest', 'LastBestBuy', 'LastBestSell',
        'RecordHigh', 'RecordLow', 'Suspending']
original_cols = ['交易日期', '契約', '到期月份(週別)', '開盤價', '最高價', '最低價', '收盤價', '成交量',
                 '漲跌價', '漲跌%', '結算價', '未沖銷契約數', '最後最佳買價', '最後最佳賣價',
                 '歷史最高價', '歷史最低價', '是否因訊息面暫停交易']
df.rename(dict(zip(original_cols, cols)), axis='columns', inplace=True)
df = df[cols]
df
```

```python
tx = df[(df['Contract']=='TX') & (df['Expiration'].apply(lambda x: len(x)==6))]
```

```python
tx.loc[:,'Date'] = pd.to_datetime(tx['Date'])
```

```python
# tx['Expiration'] = pd.to_datetime(tx['Expiration'], format='%Y%m')
```

```python
tx.set_index(['Date', 'Expiration'], inplace=True)
```

```python
tx.sort_index(inplace=True)
```

```python
num_cols = ('Volume', 'Open', 'High', 'Low', 'Close', 'OpenInterest', 'RecordHigh', 'RecordLow',
            'Change', 'Settlement', 'LastBestBuy', 'LastBestSell')
for col in num_cols:
    tx.loc[:,col] = pd.to_numeric(tx.loc[:,col], errors='coerce')
```

### prepare expiration date (2006-2016)

```python
tx = pd.read_csv('futures.csv')
tx.loc[:,'Date'] = pd.to_datetime(tx['Date'])
tx.loc[:,'Expiration'] = tx['Expiration'].astype(str)
#tx.set_index(['Date'], inplace=True)
```

```python
exps = tx.groupby('Expiration').tail(1)
#expirations = {j:i for i, j in exps[['Date', 'Expiration']] if i.strftime('%Y%m')==str(j)}
expired_only = exps[['Date', 'Expiration']].apply(lambda x: x['Date'].strftime('%Y%m')==str(x['Expiration']), axis=1)
expirations = exps[expired_only].set_index(['Expiration'])[['Date']].to_dict('index')
expirations = {i:j['Date'] for i, j in expirations.items()}
expirations
```

```python
tx1 = tx.groupby('Date').nth(0).reset_index()
tx1 = tx1.loc[:,['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Change', 'OpenInterest', 'Expiration']]
expired_only = tx1['Expiration'].apply(lambda x: x in expirations.keys())
tx1 = tx1[expired_only]
tx1['Days2Expiration'] = tx1.apply(lambda x: (expirations[x['Expiration']]-x['Date']).days, axis=1)
tx1.drop(columns=['Expiration'], inplace=True)
tx1.to_csv('tx1.csv', index=False)
```
